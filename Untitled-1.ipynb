```python
import pandas as pd
import spacy
import re
from collections import Counter

# Loading the model and text / Caricamento del modello e del testo / Загрузка модели и текста
nlp = spacy.load("en_core_web_lg")
nlp.max_length = 1500000

try:
    # Place your 'fellowship_of_the_ring.txt' file in the same directory
    with open('fellowship_of_the_ring.txt', 'r', encoding='utf-8') as file:
        text = file.read()
except FileNotFoundError:
    text = "Mr. Frodo lived in a hole. It was a great hole, but he had to go away."
    print("Attention: File not found, test text is used.")
```

```python
# Dictionary Processing / Elaborazione del dizionario/ Обработка словаря
try:
    # irst upload the file / Prima carica il file / Сначала загрузить файл
    df_cefr = pd.read_csv('english_cefr.csv')
    
    # Processing slashes / Elaborazione degli slash / Обработка слешей (civilization/civilisation)
    if 'Word' in df_cefr.columns:
        df_cefr['Word'] = df_cefr['Word'].str.split('/').explode('Word').reset_index(drop=True)
    
    # --- Now cleaning up (Fixing the bloated list) / Pulizia in corso (Correzione dell'elenco sovraccarico) / Очистка (Исправление раздутого списка) ---
    level_order = {'A1': 1, 'A2': 2, 'B1': 3, 'B2': 4, 'C1': 5, 'C2': 6}
    
    # Creating a temporary rank for sorting / Creazione di un rango temporaneo per l'ordinamento / Создаем временный ранг для сортировки
    df_cefr['Level_Rank'] = df_cefr['Level'].map(level_order)
    
    # Sorting: first by word, then by the easiest level / Ordinamento: prima per parola, poi per il livello più facile / Сортируем: сначала по слову, потом по самому легкому уровню
    df_cefr = df_cefr.sort_values(by=['Word', 'Level_Rank'])
    
    # Removing duplicates, keeping only the easiest level for each word / Rimozione dei duplicati, mantenendo solo il livello più facile per ogni parola / Удаляем дубликаты, оставляя только самый легкий уровень для каждого слова
    df_cefr = df_cefr.drop_duplicates(subset=['Word'], keep='first')
    
    # Removing the temporary column / Rimozione della colonna temporanea / Удаляем временную колонку
    df_cefr = df_cefr.drop(columns=['Level_Rank'])
    
    print(f"Dictionary loaded and optimized! Unique lemmas: {len(df_cefr)}")

except FileNotFoundError:
    print("ERROR: File 'english_cefr.csv' not found.")
```

```python
# 3. NLP Processing / Elaborazione NLP / Лингвистическая обработка
# Remove digits and clean / Rimuovi le cifre e pulisci / Удалить цифры и очистить
clean_text = re.sub(r'\d+', '', text)
doc = nlp(clean_text)

common_words = []
phrasal_verbs = []
tolkien_lore = []
rows = [] # New list for storing all data / Nuovo elenco per la memorizzazione di tutti i dati / Новый список для хранения всех данных

exclude_lore = {'mr', 'mrs', 'sir', 'st', 'great'}

for token in doc:
    if token.is_punct or token.is_space:
        continue

    # Defining basic parameters / Definizione dei parametri di base / Определяем базовые параметры
    lemma = token.lemma_.lower()
    pos = token.pos_
    is_stop = token.is_stop
    
    # 1. Processing proper nouns / Elaborazione dei nomi propri / Обработка имен собственных
    if pos == "PROPN":
        name = token.text.strip('.')
        if name.lower() not in exclude_lore:
            tolkien_lore.append(name)
            rows.append({'lemma': name.lower(), 'pos': 'PROPN', 'is_stop': False})
            continue

    # 2. Processing phrasal verbs / Elaborazione dei verbi frasali / Обработка фразовых глаголов
    if pos == "VERB":
        particles = [t.lemma_.lower() for t in token.children if t.dep_ == "prt"]
        if particles:
            phrasal = f"{lemma}_{particles[0]}"
            phrasal_verbs.append(phrasal)
            rows.append({'lemma': phrasal, 'pos': 'VERB', 'is_stop': False})
            continue

    # 3. All other words / Tutte le altre parole / Все остальные слова
    rows.append({'lemma': lemma, 'pos': pos, 'is_stop': is_stop})
    if not is_stop:
        common_words.append(lemma)

# Creating df_all with all columns at once / Creazione di df_all con tutte le colonne contemporaneamente / Создаем df_all со всеми колонками сразу
df_all = pd.DataFrame(rows)

# Merging with CEFR / Unione con CEFR / Мерджим с CEFR
df_analysis = pd.merge(df_all, df_cefr, left_on='lemma', right_on='Word', how='left')
df_analysis['Level'] = df_analysis['Level'].fillna('Not in CEFR list')
```

```python
print("\n--- 1. General Statistics / Statistiche generali / Общая статистика ---")
print(f"Total tokens for analysis: {len(df_all)}")
```

```python
print("\n--- 2. Top 30 Common Words / Top 30 parole comuni / Топ-30 обычных слов ---")
print(pd.Series(common_words).value_counts().head(30))
```

```python
print("\n--- 3. Top 20 Phrasal Verbs / Top 20 verbi frasali / Топ-20 фразовых глаголов ---")
if phrasal_verbs:
    print(pd.Series(phrasal_verbs).value_counts().head(20))
else:
    print("No phrasal verbs found.")
```

```python
print("\n--- 4. Tolkien's World (PROPN) / Il mondo di Tolkien / Мир Толкина ---")
df_lore = pd.DataFrame(tolkien_lore, columns=['name'])
print(df_lore['name'].value_counts().head(30))
```

```python
# --- 1. Preparing Categories (Names vs. Compound Words) / Preparazione delle categorie (Nomi vs Parole composte) / Подготовка категорий (Имена vs Сложные слова) ---
lore_set = set(name.lower() for name in tolkien_lore)

def categorize_extra(row):
    if row['Level'] != 'Not in CEFR list':
        return row['Level']
    if row['lemma'] in lore_set:
        return 'Tolkien Lore (Names/Places)'
    return 'Advanced / Unknown'

# Applying the Split / Applicazione della divisione / Применяем разделение
df_analysis['Level'] = df_analysis.apply(categorize_extra, axis=1)

# --- 2. Generating report / Generazione del rapporto / Генерация отчета ---
print("\n--- 5. Unique Vocabulary Analysis / Analisi del vocabolario unico / Анализ уникальных слов ---")

df_unique = df_analysis.drop_duplicates(subset=['lemma'])
unique_counts = df_unique['Level'].value_counts()
unique_percentage = (unique_counts / unique_counts.sum()) * 100

level_stats_unique = pd.DataFrame({
    'Unique Words': unique_counts,
    'Percentage (%)': unique_percentage.round(2)
})

# Sorting from Largest to Smallest / Ordinamento dal più grande al più piccolo / Сортировка от большего к меньшему
print(level_stats_unique.sort_values(by='Unique Words', ascending=False))
```

```python
print("\n--- 6. Text Density Analysis / Analisi della densità del testo / Анализ плотности текста ---")

total_counts = df_analysis['Level'].value_counts()
total_percentage = (total_counts / total_counts.sum()) * 100

level_stats_total = pd.DataFrame({
    'Total Occurrences': total_counts,
    'Percentage (%)': total_percentage.round(2)
})

# Sorting from Largest to Smallest / Ordinamento dal più grande al più piccolo / Сортировка от большего к меньшему
print(level_stats_total.sort_values(by='Total Occurrences', ascending=False))
```

```python

word_counts = df_analysis['lemma'].value_counts().reset_index()
word_counts.columns = ['lemma', 'frequency']


df_tableau = df_analysis[['lemma', 'Level', 'pos', 'is_stop']].drop_duplicates(subset=['lemma'])

df_tableau = pd.merge(df_tableau, word_counts, on='lemma', how='left')
df_tableau.to_csv('tolkien_analysis_for_tableau.csv', index=False, encoding='utf-8')

print("Success! File 'tolkien_analysis_for_tableau.csv' created")
```
